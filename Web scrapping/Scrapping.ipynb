```python
#import the library used to query a website
# import urllib2 #if you are using python3+ version, 
import urllib.request

#specify the url
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"

#Query the website and return the html to the variable 'page'
page = urllib.request.urlopen(wiki) #For python 3 use urllib.request.urlopen(wiki)

#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup

#Parse the html in the 'page' variable, and store it in Beautiful Soup format/
soup = BeautifulSoup(page, 'html.parser')

print(soup.prettify())
```

```python
# soup.<tag>: Return content between opening and closing tag including tag.
soup.script
```

```python
# soup.<tag>.string: Return string within given tag
soup.title.string
```

```python
soup.a
soup.find_all("a")
```

```python
all_table = soup.find_all('table')
```

```python
right_table=soup.find('table', class_='wikitable sortable plainrowheaders')
right_table
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') #To store second column data
    if len(cells)==6: #Only extract table body not heading
        import antigravity
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))

#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(A,columns=['Number'])
df['State/UT']=B
df['Admin_Capital']=C
df['Legislative_Capital']=D
df['Judiciary_Capital']=E
df['Year_Capital']=F
df['Former_Capital']=G
df

# You can also look at the other attributes like .parent, .contents, .descendants and .next_sibling, 
# .prev_sibling and various attributes to navigate using tag name.
```

```python
css_soup = BeautifulSoup('<p class="body"></p>', 'html.parser')
css_soup.p['class']
# ['body']

css_soup = BeautifulSoup('<p class="body strikeout"></p>', 'html.parser')
css_soup.p['class']
# ['body', 'strikeout']
```

```python
id_soup = BeautifulSoup('<p id="my id"></p>', 'html.parser')
id_soup.p['id']
# 'my id'
```

```python
rel_soup = BeautifulSoup('<p>Back to the <a rel="index">homepage</a></p>', 'html.parser')
rel_soup.a['rel']
# ['index']
rel_soup.a['rel'] = ['index', 'contents']
print(rel_soup.p)
# <p>Back to the <a rel="index contents">homepage</a></p>
```

```python
no_list_soup = BeautifulSoup('<p class="body strikeout"></p>', 'html.parser', multi_valued_attributes=None)
no = BeautifulSoup('<p class="body strikeout"></p>', 'html.parser')
no_list_soup.p['class'] = ['body','strike']
no.p['class'] = ['body','strike'] 
print(no_list_soup.p['class'])
print(no.p['class'])
# 'body strikeout'
```
